```python
# list comprehensions
# it provides a concise way to create lists
# they are more readable and efficient way than than for loops
# Syntax:
#  [ expression for item in iterable if condition]
```

```python
# create a iist of squares
squres = [x**2 for x in range (1,6)]
print(squres)
```

```python
# filter even numbers
even = [x for x in range (10) if x%2==0]
print(even)
```

```python
# convert string to uppercase
names = ["sam", "rohit", "ankit"]
uppercase_names = [name.upper() for name in names]
print(uppercase_names)
```

```python
# flatten a nested list
matrix = [[1,2], [3,4], [5,6]]
flattened = [num for sublist in matrix for num in sublist]
print(flattened)
```

```python
# generate a list of tuples
pairs = [(x, y) for x in range (3) for y in range (3)]
print(pairs)
```

```python
# above questions without creating list comprehensions
squares = []
for x in range (1,6):
    squares.append(x**2)
print(squares)
```

```python
even = []
for x in range (10):
    if x%2==0:
        even.append(x)
print(even)
```

```python
names = ["sam", "rohit", "ankit"]
uppercase_names = []
for name in names:
    uppercase_names.append(name.upper())
print(uppercase_names)
```

```python
matrix = [[1,2], [3,4], [5,6]]
flattened = []

for sublist in matrix:
  for num in sublist:
    flattened.append(num)

print(flattened)
```

```python
pairs = []
for x in range (3):
 for y in range (3):
  pairs.append((x,y))
print(pairs)
```

dictionary comprehensions

```python
# it offers aconcise way to create dictionaries from iterables
# similar to list comprehensions, they allow you ti construct dictionaries in a
# single line with an optimal condition for filtering

# syntax:
#  {key expression : value expression for expression in iterable if condition}
```

```python
# create a dictionary of squares
squares = {x: x**2 for x in range (1,8)}
print(squares)
```

```python
# filter even numbers
even = {x: x**2 for x in range (10) if x % 2 == 0}
print(even)
```

```python
# reverese a dictionary
orignal = {'a' : 1, 'b' : 2, 'c' : 3}
rev = {v : k for k, v in orignal.items()}
print(rev)
```

```python
# create dictionary with word and number of letters in that word from the list
words = ["apple", "banana", "cherry"]
word_lengths = {word: len(word) for word in words}
print(word_lengths)
```

```python
# create a dictionary with number and and its value be even or odd using comprehension
num = {x: ("even" if x%2==0 else "odd") for x in range (1,11)}
print(num)
```
