```python
df = pd.read_csv('/content/covid_toy.csv')
df
```

```python
x = df.drop(columns = ['has_covid'])
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
titanic_df = pd.read_csv('/content/titanic.csv')
titanic_df
```

```python
titanic_df.isnull().sum()
```

```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Fare'] = titanic_df['Fare'].fillna(titanic_df['Fare'].mean())

# titanic_df
```

```python
titanic_df = titanic_df.drop(columns = ['Cabin'])
# titanic_df
```

```python
titanic_df.isnull().sum()
```

```python
x = titanic_df.drop(columns = ['Survived'])
# x
```

```python
y = titanic_df['Survived']
# y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=20)
```

```python
print("Total data shape :",titanic_df.shape)
print("---------------------------")
print("Input data(x) shape :",x.shape)
print("x_train data shape :",x_train.shape)
print("x_test data shape :",x_test.shape)
print("-------------------------------")

print("Target data shape (y):",y.shape)
print("y_train data shape :",y_train.shape)
print("y_test data shape :",y_test.shape)
print("-------------------------------")
```
