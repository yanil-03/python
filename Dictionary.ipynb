### Functions in dictionary

```python
a = {'Brand': 'Kia', 'Model': 'Seltos', 'Year': 2024}
print(a)
# 1. len(a)
len(a)
```

```python
a = {'Brand': 'Kia', 'Model': 'Seltos', 'Year': 2024}
print(a)
# 2. pop()
a.pop('Model')
print(a)
```

```python
# 3. popitem()
a = {'Brand': 'Kia', 'Model': 'Seltos', 'Year': 2024, 'Color': 'white'}
print(a)
a.popitem()
print(a)
```

```python
# 4. clear()
a = {'Brand': 'Kia', 'Model': 'Seltos', 'Year': 2024, 'Color': 'white'}
print(a)
a.clear()
print(a)
```

```python
Raj = {"Jackets":500, "Jeans":1500, "Shoes":8000}
print(Raj)
a = []
for i in Raj.values():
  a.append(i)

sum = 0
for i in a:
  sum += i
print("Your total bill is",sum)

b = []
for i in Raj.keys():
  b.append(i)
print("Your item list is :",b)
```

```python
# from "apple" create a dictionary where keys are characters and values are
# how many times they occur

s = "apple"
d = {}
for i in s:
  if i in d:
    d[i] += 1
  else:
    d[i] = 1
  # print(d)
print(d)
```

```python
# merge {'a': 1, 'b': 2} and {'b': 3, 'c': 4}. if key is same then add the values.
d1 = {'a': 1, 'b': 2}
d2 = {'b': 3, 'c': 4}
result = {}

for k in d1:
  result[k] = d1[k]

for k in d2:
  if k in result:
    result[k] += d2[k]
  else:
    result[k] = d2[k]

print(result)
```

```python
# from {'a': 1, 'b': 2, 'c': 1}, amke {1: ['a', 'c'], 2; ['b']}

d1 = {'a': 1, 'b': 2, 'c': 1}
inv = {}

for k in d1:
  val = d1[k]
  if val in inv:
    inv[val] += [k]
  else:
    inv[val] = [k]

print(inv)
```
