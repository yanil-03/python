```python
# write a program to calculate sum of cube of digits of a givem number
i=123
sum=0
while (i>0):
    sum=sum + (i%10)*(i%10)*(i%10)
    i = i//10
print(sum)

# 1**3 + 2**3 + 3**3 = 1+8+27=36
```

```python
# Armstrong number : sum of cube of digits of a givem number == orignal number
  #  153 = 1**3 + 5**3 + 3**3 =153


i=153
orig=i
sum=0
while (i>0):
    sum=sum + (i%10)*(i%10)*(i%10)
    i = i//10
if(sum==orig):
    print("armstrong number")
else:
    print("not armstrong number")
```

```python
# sum of even and odd digits of a given number

i =1234
odd =0
even =0
while(i>0):
  d=i%10
  if(d%2==0):
    even=even+d
  else:
    odd=odd+d
  i=i//10
print("Sum of odd digits: ",odd)
print("Sum of even digits: ",even)
```

```python
i=int(input("Enter a number: "))
orig = i
rev = 0
while(i>0):
  rev = (rev*10) + (i%10)
  i = i//10

if(orig==rev):
  print("palindrome number")
else:
  print("not palindrome number")
```

```python
#Ques: write a program to count the number of vowels in a given a given string using for loop
text = "Hello World"
vowel = "aeiouAEIOU"
count = 0
cons = 0
for char in text:          #
  if char in vowel:
    count+=1
  else:
    cons+=1
print("Number of vowels:",count)
print("Number of consonants:",cons)
```
