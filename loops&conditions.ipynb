```python
# program to calculate bill for 5 items and if total price exceexds 100,
# a discount of 10% ios applied

total_price = 0
for i in range(5):
  price = float(input(f"Enter the price of item {i + 1}: "))
  total_price += price

  if total_price > 100:
    total_price *= 0.9
print(f"Total bill after discount, if any ${total_price:.2f}")
```

```python
# program of grading system for "pass, "fail" or excellent(fail- <40,pass: 40-80,excellent: >80)

marks = int(input("Enter your marks: "))
if (marks > 0 and marks <= 40):
  print("Fail")

elif (marks > 40 and marks <= 80):
  print("Pass")

elif (marks > 80 and marks <= 100):
  print("Excellent")

else:
  print("invalid marks!!!")
```

```python
# program that keeps asking user to input a password
# until it is 8 character long. if less program should ask again for password

while True:
  password = input("Enter a password: ")

  if len(password) >= 8:
    print("Password accepted")
    break

  else:
    print("Password should be 8 characters long")
```

```python
# create a list to store daily tasks. allow the user to add 5 tasks one by one,
# then display it all.

tasks = []
for i in range(5):
  task = input(f"Enter task {i + 1}: ")
  tasks.append(task)

print("Your task for today are:")
for task in tasks:
  print(f"- {task}")
```

```python
# program to swap values of two variables

a = 10
b = 20
print(f"Before swapping: a = {a}, b = {b}")
a, b = b, a
print(f"After swapping: a = {a}, b = {b}")
```

```python
# program to filter out odd and even numbers from a given list
numbrs = [10, 25, 36, 47, 52, 63, 78]
even_numbers = []
odd_numbers = []
for num in numbrs:
  if num % 2 == 0:
    even_numbers.append(num)
  else:
    odd_numbers.append(num)

print("Odd numbers: ", odd_numbers)
print("Even numbers: ", even_numbers)
```

```python
# you have list of prices , and you need to categorize each product
# based on its price
prices = [1200, 800, 450, 300, 950, 2000]
categories = []
for price in prices:
  if price > 1000:
    categories.append("Expensive")
  elif 500 <= price <= 1000:
    categories.append("Moderate")
  else:
    categories.append("Cheap")

print("Categories: ",categories)
```

```python
# count number of vowels in a sentance
sentance = "Hello, how are you?"
vowels = "aeiouAEIOU"
count = 0
for char in sentance:
  if char in vowels:
    count += 1
print("Number of vowels: ", count)
```

```python
# check palindrone words in a list
words = ["madam", "hello", "level", "python", "racecar" ]
palindrome = []
for word in words:
  if word == word[::-1]:
    palindrome.append("True")
  else:
    palindrome.append("False")
print(palindrome)
```

```python
# categorize tempratture
# # cold-> t<15
# warm-> 15<=t<=30
# hot-> t>30

temperature = [10, 22, 35, 28, 15, 5, 40]
categories = []
for temp in temperature:
  if temp < 15:
    categories.append("Cold")
  elif 15 <= temp <= 30:
    categories.append("Warm")
  else:
    categories.append("Hot")

print("Categories: ",categories)
```
