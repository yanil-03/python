```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/dsjob.csv')
```

```python
from sklearn.impute import SimpleImputer
```

```python
# si_mean = SimpleImputer(strategy='mean')
si_frequent = SimpleImputer(strategy='most_frequent')
```

```python
df['gender'] = si_frequent.fit_transform(df[['gender']]).ravel()
df['enrolled_university'] = si_frequent.fit_transform(df[['enrolled_university']]).ravel()
df['education_level'] = si_frequent.fit_transform(df[['education_level']]).ravel()
df['major_discipline'] = si_frequent.fit_transform(df[['major_discipline']]).ravel()
df['experience'] = si_frequent.fit_transform(df[['experience']]).ravel()
df['company_size'] = si_frequent.fit_transform(df[['company_size']]).ravel()
df['company_type'] = si_frequent.fit_transform(df[['company_type']]).ravel()
df['last_new_job'] = si_frequent.fit_transform(df[['last_new_job']]).ravel()
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
lb = LabelEncoder()
oe = OrdinalEncoder(categories = [['Male', 'Female','Other']])
ohe = OneHotEncoder(drop = 'first', sparse_output = False)
```

```python
df['city'] = lb.fit_transform(df['city'])        #label encoding for city sub categories
df['education_level'] = lb.fit_transform(df['education_level'])        #label encoding for city sub categories
df['major_discipline'] = lb.fit_transform(df['major_discipline'])        #label encoding for city sub categories
df['company_type'] = lb.fit_transform(df['company_type'])        #label encoding for city sub categories
df['gender'] = oe.fit_transform(df[['gender']])   #ordinal encoding for gender sub categories
```

```python
df['relevent_experience'] = ohe.fit_transform(df[['relevent_experience']])  #one hot encoding for relevent experience
```

```python
df = pd.get_dummies(df , columns = ['enrolled_university'], drop_first = True)
```

```python
df['experience'] = lb.fit_transform(df['experience'])         #label encoding for experience sub categories
```

```python
df['company_size'] = lb.fit_transform(df['company_size'])         #label encoding for comapny size sub categories
```
