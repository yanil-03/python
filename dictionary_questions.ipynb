```python
# Create a simple grocery billing system. The user adds items from a predefined list of products. At the
# end, the system generates a bill with total amount, taxes (e.g. 5%), and a discount if applicable (e.g.
# 10% off on total above ₹1000).

# input:
# products = {
#     "apple" : 100,
#     "banana" : 40,
#     "milk" : 60,
#     "bread" : 50,
#     "eggs" : 70

# }

# A list of tuples (item name, quantity) entered by the user.
# Example:
# cart = [("apple", 2), ("milk", 1), ("bread", 3)]

# OUTPUT:-
# Print a formatted bill showing:
# o Item-wise cost
# o Subtotal
# o Tax (5%)
# o Discount (10% if total > ₹1000)
# o Final amount to be paid
```

```python
products = {
    "apple" : 100,
    "banana" : 40,
    "milk" : 60,
    "bread" : 50,
    "eggs" : 70

}
cart = []
total_cost = 0

print(products)

while True:
    item = input("Enter the item name (or 'exit' to finish): ")
    if item == "exit":
        break

    quantity = int(input("Enter the quantity: "))
    cart.append((item, quantity))

if not cart:
    print("Your cart is empty.")
else:
  for item, quantity in cart:

    if item in products:
      print(f"\n{item}: {products[item]} X {quantity} = Rs.{products[item] * quantity}")
      total_cost += products[item] * quantity

    else:
      print(f"\n{item} is not available in the store.")
# print(cart)
print(f"\n\nSubtotal: Rs.{total_cost}")

total =total_cost + (total_cost * 0.05)
print(f"\nTax (5%): Rs.{total}")

if(total > 1000):
  discount = total * 0.1
  print(f"\nDiscount (10%): Rs.{discount}")
  final_amount = total - discount
else:
  final_amount = total

print(f"\nFinal amount to be paid: Rs.{final_amount}")
```

```python
# Problem Statement:
# Build a program for a school to calculate each student’s average score, assign grades, and summarize
# how many students got each grade.

# input: a dictionary of students and their marks as a list
# students = {
#     "Alice": [85, 92, 78],
#     "Bob": [70, 60, 65],
#     "Charlie": [90, 95, 88],
#     "David": [55, 60, 52]
# }
# OUTPUT:
# for each students:
#   print name, average marks, Grade
#   (e.g. A: 90+, B: 80-89, C: 70-79, D: 60-69, F: < 60)

# Finally:
#  Print a summary count of students in each grade category.
```

```python
students = {
    "Alice": [85, 92, 78],
    "Bob": [70, 60, 65],
    "Charlie": [90, 95, 88],
    "David": [55, 60, 52]
}


# for name, scores in students.items():
#   print(f"\n{name}: {scores}")
#   total_marks = sum(scores)
#   average_score = total_marks / len(scores)
#   print(f"Average Score: {average_score}")


grade_counts = {"A": 0, "B": 0, "C": 0, "D": 0, "F": 0}

for name, scores in students.items():
  average_score = sum(scores) / len(scores)
  print(f"{name}: Average Score = {average_score}", end=" ")

  if average_score >= 90:
    grade = "A"
  elif average_score >= 80:
    grade = "B"
  elif average_score >= 70:
    grade = "C"
  elif average_score >= 60:
    grade = "D"
  else:
    grade = "F"

  print(f"Grade = {grade}")
  grade_counts[grade] += 1

print("\nGrade Summary:")
for grade, count in grade_counts.items():
  print(f"{grade}: {count} students")

```
