```python
# Q.1 Case Study 1: Grocery Billing System
# Problem Statement:
# Create a simple grocery billing system. The user adds items from a predefined list of products. At the
# end, the system generates a bill with total amount, taxes (e.g. 5%), and a discount if applicable (e.g.
# 10% off on total above ₹1000).

# Input:
# A dictionary of items with their prices.
# Example:

# products = {
#   "apple": 100,
#   "banana": 40,
#   "milk": 60,
#   "bread": 50,
#   "eggs": 70
# }

# A list of tuples (item name, quantity) entered by the user.
# Example:
# cart = [("apple", 2), ("milk", 1), ("bread", 3)]

# Output:

# Print a formatted bill showing:
# o Item-wise cost
# o Subtotal
# o Tax (5%)
# o Discount (10% if total > ₹1000)
# o Final amount to be paid
```

```python
products = {
  "apple": 100,
  "banana": 40,
  "milk": 60,
  "bread": 50,
  "eggs": 70
}
subtotal = 0
tax = 0.05
discount = 0.1

cart = []

def print_dict(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:\tRs.{value}")

print_dict(**products)
while True:
    item = input("Enter the item name (or type exit to finish): ")

    if item == "exit":
        break
    elif item not in products:
        print("Invalid item name. Please try again.")
        continue

    quantity = int(input("Enter the quantity: "))
    cart.append((item, quantity))

print("\n\nYOUR CART :-")

if not cart:
   print("Your cart is empty.")
else:
  for items, quantity in cart:
    print(f"{items}:\t{quantity} X {products[items]}\t\tRs.{products[items]*quantity}")
    subtotal += products[items]*quantity

  print("\n")
  # BILLING
  # print(cart)
  print(f"Subtotal:\t\t\tRs.{subtotal}")
  print(f"Tax (5%):\t\t\tRs.{subtotal*tax}")

  total = subtotal + (subtotal*tax)
  if total > 1000:
    print(f"Discount (10%):\t\t\tRs.{total*discount}")
    total -= total*discount

  print(f"Total amount payable:\t\tRs.{total}")
```

```python
# Problem Statement:
# Build a program for a school to calculate each student’s average score, assign grades, and summarize
# how many students got each grade.

# Input:
# A dictionary of students and their marks (as a list).
# Example:

# students = {
#   "Alice": [85, 92, 78],
#   "Bob": [70, 60, 65],
#   "Charlie": [90, 95, 88],
#   "David": [55, 60, 52]
# }

# Output:

# For each student:
#  Print Name, Average Marks, Grade
#  (e.g. A: 90+, B: 80-89, C: 70-79, D: 60-69, F: < 60)

# Finally:
#  Print a summary count of students in each grade category.
```

```python
students = {
  "Alice": [85, 92, 78],
  "Bob": [70, 60, 65],
  "Charlie": [90, 95, 88],
  "David": [55, 60, 52]
}
grade_count = {"A": 0, "B": 0, "C": 0, "D": 0, "F": 0}

def grading(n):
  if n >= 90:
    return "A"
  elif 89 >= n >= 80:
    return "B"
  elif 79 > n >= 70:
    return "C"
  elif 69 >= n >= 60:
    return "D"
  else:
    return "F"



for name, marks in students.items():
  avg = sum(marks)/len(marks)
  if grading(avg) in grade_count:
    grade_count[grading(avg)] += 1

  print(f"\n\nStudent Name:\t{name}\nAverage Marks:\t{avg}\nGrade:\t{grading(avg)}")


print("\nSummary count of students in each grade category : - \n")
for grade, count in grade_count.items():
  print(f"{grade}:\t{count}")
```
