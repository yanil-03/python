```python
# Q.1 Reverse a list without using built-in methods.

# Input: [10, 20, 30, 40]
# Output: [40, 30, 20, 10]
a = [10, 20, 30, 40]
rev = []
for i in range(len(a)-1,-1,-1):
  rev.append(a[i])
print(rev)
```

```python
# Q.2 Create a user defiend list and calculate sum of all elements that present in list .
a = []
sum = 0
limit = int(input("Enter the limit: "))
for i in range (limit):
  a.append(input("Enter elements: "))
print(a)
for i in range(len(a)):
  sum = sum + int(a[i])
print(sum)
```

```python
# Q.3 Remove all duplicates from a list while maintaining order.

# # Input: [1, 2, 2, 3, 4, 3]
# # Output: [1, 2, 3, 4]

a = [1, 2, 2, 3, 4, 3]
b = []
for i in a:
  if i in b:
    continue
  else:
    b.append(i)
print(b)
```

```python
# Q.5 Convert a list of tuples into a dictionary.

# # Input: [("a", 1), ("b", 2), ("c", 3)]
# # Output: {'a': 1, 'b': 2, 'c': 3}
a = [("a", 1), ("b", 2), ("c", 3)]
b = {}
for i in a:
  b[i[0]] = i[1]
print(b)
```

```python
# Q.6 Write a function that takes a tuple of integers and returns a tuple with only even numbers.
# # Input: (1, 2, 3, 4, 5, 6)
# # Output: (2, 4, 6)
a = (1, 2, 3, 4, 5, 6)
b=[]
a = list(a)
for i in a:
  if i%2 == 0:
    b.append(i)
b = tuple(b)
print(b)
```

```python
# Q.7 Find common elements between two tuples.
# # Input: (1, 2, 3), (2, 3, 4)
# # Output: (2, 3)
a = (1, 2, 3)
b = (2, 3, 4)
c = []
a = list(a)
b = list(b)

for i in a:
  if i in b:
    c.append(i)

c = tuple(c)
print(c)
```

```python
# Q.8 Merge two dictionaries and sum values of common keys.
# # Input: {'a': 10, 'b': 20}, {'a': 5, 'c': 7}
# # Output: {'a': 15, 'b': 20, 'c': 7}
a = {'a': 10, 'b': 20}
b = {'a': 5, 'c': 7}
c = {}
for k in a:
  c[k] = a[k]

for k in b:
  if k in c:
    c[k] += b[k]
  else:
    c[k] = b[k]


print(c)
```
