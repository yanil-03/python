## SETS IN PYTHON


```python
# 1. Introduction = {element1,element2,element3......element n}
# 2. mutable datatype---> we can change or update values at runtime
# 3. It will have only unique values which means it cannot contains duplicate values
# 4. order of values is not fixed
```

```python
a = {'sam', 45 , 'sam' , 'raj' , 'sam' , 67 , 90 , 45 , 45}
type(a)
print(a)
```

```python
# MUTABLITY IN SETS
a = {'sam', 45 , 'sam' , 'raj' , 'sam' }
print(a)
a.add('ram')
print(a)
a.remove('sam')
print(a)
```

```python
# LOOPS IN SETS
a = {'sam', 45 , 'sam' , 'raj' , 'sam' }
for i in a:
  print(i)
```

```python
a = [1,5,2,4,8,3,9,34,10,2,2,5,7,8,1]
print(a)
b = set(a)
print(b)
c = list(b)
c.sort()
print(c)
print('second minimum :',c[1])
print('second maximum :',c[-2])
```

```python
# SET OPERATIONS
a = {1,2,3}
b = {4,5,6}
c = a + b
c
```

```python
a = {1,2,3,4}
b = {3,4,5,6}
c = a.union(b)
print(c)
e = a|b             ###second method of union
print(e)
d = a.intersection(b)
print(d)
f = a & b           ###second method of intersection
print(f)
```
