```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# label encoding

df_lb = df.copy()
```

```python
lb = LabelEncoder()
df_lb['Gender'] = lb.fit_transform(df_lb['Gender'])
```

```python
df_lb.sample(5)
```

```python
# Oridinal encoding
df_oe = df.copy()
```

```python
df_oe['Gender'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = [['Female', 'Male']])
```

```python
df_oe['Gender'] = oe.fit_transform(df_oe[['Gender']])
```

```python
df_oe
```

```python
# OneHot encoding
df_ohe = df.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop= 'first'  ,sparse_output= False)
```

```python
df_ohe['Gender'] = ohe.fit_transform(df_ohe[['Gender']])
```

```python
df_ohe
```

```python
# get dummies
df_dum = df.copy()
```

```python
df_dum = pd.get_dummies(df_dum, drop_first= True, columns= ['Gender'])
df_dum
```

```python
# STANDARDIZATION
df_std = df.copy()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df_std['User ID'] = lb.fit_transform(df_std['User ID'])
df_std['Gender'] = lb.fit_transform(df_std['Gender'])
```

```python
sc = StandardScaler()
```

```python
np.round(df_std.describe(), 2)
```

```python
df_sc = sc.fit_transform(df_std)
```

```python
df_new = pd.DataFrame(df_sc, columns= df_std.columns)
```

```python
np.round(df_new.describe(), 2)
```

```python
# NORMALIZATION
df_norm = df.copy()
```

```python
df_norm['User ID'] = lb.fit_transform(df_norm['User ID'])
df_norm['Gender'] = lb.fit_transform(df_norm['Gender'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn = MinMaxScaler()
```

```python
np.round(df_norm.describe(), 2)
```

```python
df_mn = mn.fit_transform(df_norm)
```

```python
df_mn_new = pd.DataFrame(df_mn, columns= df_norm.columns)
```

```python
np.round(df_mn_new.describe(), 2)
```
