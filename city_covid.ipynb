```python
import pandas as pd
```

```python
df = pd.read_csv('/content/covid_toy.csv')
df
```

```python
g = df.filter(items = ['city' , 'has_covid'])
g

```

```python
cty = df.groupby('city')
cty.groups
```

```python
kolkata = cty.get_group('Kolkata')
kolkata = kolkata.rename(columns = {'has_covid' : 'Kolkata'})
kolkata = kolkata.drop('city' , axis = 1)
kolkata
```

```python
# bangalore
bangalore = cty.get_group('Bangalore')
bangalore = bangalore.rename(columns = {'has_covid' : 'Bangalore'})
bangalore = bangalore.drop('city' , axis = 1)
bangalore
```

```python
# delhi
delhi = cty.get_group('Delhi')
delhi = delhi.rename(columns = {'has_covid' : 'Delhi'})
delhi = delhi.drop('city' , axis = 1)
delhi
```

```python
mumbai = cty.get_group('Mumbai')
mumbai = mumbai.rename(columns = {'has_covid' : 'Mumbai'})
mumbai = mumbai.drop('city' , axis = 1)
mumbai
```

```python
new_df1 = pd.concat([kolkata , bangalore , delhi , mumbai] , axis = 0)
# new_df1 = new_df1.dropna(axis = 0)
new_df1
```
