```python
# 1. zeros() -> it will create an array in which all elements are zero either in
# one dimentional or multi dimensional
import numpy as np
a = np.zeros(4)
print(a)
a = np.zeros((3,4))
print(f"\n{a}\n")
b = np.zeros((3,4,5))
b
```

```python
# 2. ones() -> it will create an array in which all elements are 1 either in
# one dimentional or multi dimensional
a = np.ones(3)
print(a)
a = np.ones((3,4))
a
```

```python
# 3. eye() -> it will create an array in ehich all the diagonal positional
#  elements are 1 and rest are 0
a = np.eye(3)
a
```

```python
# 4. diag() -> it will customize the diagonal positional elements and  rest are 0
a = [23,54,1,6]
b = np.diag(a)
b
```

```python
# RANDOM MODULE
  # a. randint() ---> it will create random numbers in a given range
  # syntax - np.random.randint(start, end, total_values)
a= np.random.randint(1,10,3)
a
```

```python
# b. rand()---> it will create an array in which all the values will be in range
#  of 0 to 1
a = np.random.rand(3)
a
```

```python
# c. seed() ---> it will fix the random generated values
np.random.seed(10)
a = np.random.randint(1,10,3)
a
```

```python
# VIEW VS COPY
# view means modification in original data
# copy means modification in duplicate data which will not reflect in the orignal
# data
a = np.array([1,2,3,4,5])
print(a)
b = a[1:4]
b[:] = 0

print(b)
print(a)
```

```python
a = np.array([1,2,3,4,5])
print(a)
b = a[1:4].copy()
b[:] = 0

print(b)
print(a)
```
