```python
# Exception handling
# 1. try block ---> code pass who can execute
# 2. Exception ---> message pass when we show error in our existing code
```

```python
try:
  a = 5
  b = a/2
  print(b)
  b = a/0
  print(b)
except Exception as e:
  print("please check your value",e)
```

```python
def exception_handling():
  try:
    result = 10/0
    print(result)
  except ZeroDivisionError as e:
    print("you cannot divide by zero",e)
exception_handling()
```

```python
def exception_handling():
  try:
    number = int(input("enter a number:"))
    print(number)
  except ValueError as e:
    print("you cannot enter characters only pass integers",e)
exception_handling()
```

```python
def exception_handling():
  try:
    list_data = [1,2,3]
    print(list_data[2])
    print(list_data[6])
  except IndexError as e:
    print("Index error",e)
exception_handling()
```

```python
def exception_handling():
  try:
    data = {"name":"ram"}
    print(data["name"])
    print(data["subject"])
  except KeyError as e:
    print("Caught key Error",e)
exception_handling()
```
