```python
# a company tracks record of 6 employees in a list
# task is to determine which employee achieved the sales above 5000
# sales = [4500, 6000, 3000, 8000, 5500, 4200]
```

```python
sales = [4500, 6000, 3000, 8000, 5500, 4200]
performers = []

for i in range(len(sales)):
    if sales[i] > 5000:
        performers.append(f"Employee {i+1}")

print(performers)
```

```python
# find the most expensive product in a store
# products=["laptop", "phone", "headphones", "monitor", "keyboard"]
# prices = [800, 500, 200, 300, 100]
```

```python
products=["laptop", "phone", "headphones", "monitor", "keyboard"]
prices = [800, 500, 200, 300, 100]
max_price = max(prices)
index = prices.index(max_price)
print(f"{products[index]} - ${max_price}")
```

```python
products=["laptop", "phone", "headphones", "monitor", "keyboard"]
prices = [800, 500, 200, 300, 100]
min_price = min(prices)
index = prices.index(min_price)
print(f"{products[index]} - ${min_price}")
```

```python
products = ["Shirts", "Jeans", "Jackets", "Shoes", "Hats"]
stock = [3, 10, 4, 15, 7]
low_stock = []
for i in range(len(products)):

  if stock[i] < 5:
    index = stock.index(stock[i])
    low_stock.append(products[index])
print(low_stock)
```

```python
# calculate employee bonuses based on performence
# scored above 80 receive $500
# scored b/w 50 and 80 gets $200
# others get no bonus
```

```python
employees = ["Aice", "Bob", "Charlie", "David", "Eve"]
scores = [85, 70, 90, 45, 60]
bonuses = []
for i in range(len(scores)):
  if scores[i] > 80:
    bonus = 500
  elif 50 <= scores[i] <= 80:
    bonus = 200
  else:
    bonus = 0
  bonuses.append(f"{employees[i]} gets ${bonus}")
# print(bonus)
print(", ".join(bonuses))
```
