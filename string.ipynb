```python
# string is a imutable datatype. imutable means we cannot change or update values at runtime
# indexing starts from 0 and length starts from 1
# forward indexing starts from 0 and backward indexing starts from -1
```

```python
#forward indexing
a = "i love maths"
print(a[0])
print(a[1])
print(a[2])
print(a[3])
print(a[4])
print(a[5])
```

```python
#backward indexing
a = "i love maths"
print(a[-1])
print(a[-2])
print(a[-3])
print(a[-4])
print(a[-5])
```

```python
# SLICING IN STRIMG

# str[start_index_value:last_index_value:step value]
a = "i love maths"
print(a[2:7])
print(a[7:12])
print(a[:7])
print(a[7:])
print(a[:])
print(a[0:11:2])
```

```python
# STRING FUNCTIONS
#1. lower() and upper() :- convert the string to lowercase or uppercase
a="Help"
print(a.lower())
print(a.upper())
```

```python
# 2. srip() :- remove whitespaces from both ends of the string
txt = "    Hello world     "
print(txt.strip())
```

```python
# 2. lsrip() :- remove whitespaces from left ends of the string
txt = "    Hello world     "
txt.lstrip()
```

```python
# 2. rsrip() :- remove whitespaces from right ends of the string
txt = "    Hello world     "
txt.rstrip()
```

```python
# 3.replace():-  replaces the substring with another string
text = "ï love python"
print(text.replace("python","coding"))
```

```python
# 4. split() :- split the string into list
text= "apple, banana, cherry"
text.split(", ")
```

```python
# 5.  join() :- join the list into string
text=['apple', 'banana', 'cherry']
print(", ".join(text))
```
