### *args(non-keyword arguments)

```python
def add_num(*args):
  total = sum(args)
  print(f"Sum: {total}")

add_num(1,2,3)
#
```

# **kwargs(keyword arguments)

```python
def print_details(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

print_details(name="John", age=25, city="New York")
```

# using *args and **kwargs together

```python
def print_all_details(*args, **kwargs):
  print(f"Positional Arguments: {args}")
  print(f"Keyword Arguments: {kwargs}")

print_all_details(1, 2, 3, name="John", age=25)
```

```python
# find maximun number using *args
def find_max(*args):
  return max(args)

print(find_max(10, 5, 25, 8, 100))
print(find_max(-1, -10, -5, 0))
```

```python
# calculate the product using args
def find_product(*args):
  result = 1
  for num in args:
    result *= num
  return result

print(find_product(2,3,4))
print(find_product(5, 5))
```

# assignment

```python
# create a function to to find the sum of digits of a number
def sum_of_digits(a):
  sum = 0
  while(a>0):
    sum += (a%10)
    a = a//10
  return sum

num = int(input("Enter a number: "))
print("Sum of digits : ",sum_of_digits(num))
```

```python
# create a function to to find the sum of squares of digits of a number
def sum_sq_digits(a):
  sum = 0
  while(a>0):
    sum += ((a%10)**2)
    a = a//10
  return sum

num = int(input("Enter a number: "))
print("Sum of squares of digits : ",sum_sq_digits(num))
```

```python
# create a function to to find the sum of cubes of digits of a number
def sum_cube_digits(a):
  sum = 0
  while(a>0):
    sum += ((a%10)**3)
    a = a//10
  return sum

num = int(input("Enter a number: "))
print("Sum of cubes of digits : ",sum_cube_digits(num))
```
